```python
#1. Create a 1D array with values from 0 to 20 using np.arange.
arr1=np.arange(0,21)
arr1
```

```python
#2. Create an array of 50 zeros and change the 10th value to 5.
arr2=np.zeros(50)
arr2[9]=5
arr2
```

```python
#3. Create an array of 100 ones and reshape it to 25×4.
arr3=np.ones(100).reshape(25,4)
arr3
```

```python
#4. Create an array of 10 evenly spaced values between 1 and 5 using np.linspace.
arr4=np.linspace(1,5,10)
arr4
```

```python
#5. Create a 3×3 identity matrix using NumPy.
arr5=np.identity(3)
arr5
```

```python
#6. Create a 4×4 array filled with the value 7.
# np.full(shape,value): It will fill the array with the given value
arr6=np.full((4,4),7)
arr6
```

```python
#7. Create a random integer array of size 5×5 with values between 10 and 50.
arr7=np.random.randint(10,50,25).reshape(5,5)     # or we can write np.random.randint(10,50,size=(5,5))
arr7
```

```python
#8. Create a 1D array of 20 random normal values using np.random.randn().
arr8=np.random.randn(20)
arr8
```

```python
#9. Given an array arr, print its dimensions, shape, size, and datatype
arr9=np.array([10,20,30,40,50])
print(arr9.ndim)
print(arr9.shape)
print(arr9.size)
print(type(arr9))
```

```python
#10. Convert a 1D array of length 9 into a 3×3 matrix.
arr10=np.arange(1,10).reshape(3,3)
arr10
```

```python
#11. Convert a 3×4 matrix into a 1×12 array using reshape.
arr11=np.arange(1,13).reshape(3,4)
print(arr11)
arr11.reshape(1,12)
```

```python
#12. Create a 3D array of shape (2,3,4) and print its shape.
arr=np.arange(1,25).reshape(2,3,4)
arr.shape
```

```python
#13. How do you check if an array is C-contiguous or F-contiguous?
arr13=np.arange(1,10)
b=arr13.reshape(3,3)
b.flags
c=np.asfortranarray(b)
c.flags
```
