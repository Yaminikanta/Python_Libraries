```python
# WAF(Write A Function) to find prime numbers between 1 to 100

l1=[]
def find_primes():
    for num in range(2, 101):                       # numbers from 2 to 100
        for i in range(2, int(num ** 0.5) + 1):     # check divisibility up to sqrt(num)
            if num % i == 0:
                break
        else:
            l1.append(num)    


find_primes()                                       # call the function
print(l1)
print(type(l1))

length=len(l1)
print(length)

arr6=np.array(l1)
print(arr6)

print(type(arr6))

arr6.reshape(5,5)
```

### OPERATORS:-

##### ARITHMETIC OPERATORS

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
arr1+arr2 #array([ 7,  9, 11, 13, 15])
np.add(arr1,arr2) #array([ 7,  9, 11, 13, 15])
```

```python
# np.add(op1,op2): it will add two arrays w.r.to index


arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
print(arr1-arr2)                                      #array([-5, -5, -5, -5, -5])
print(arr2-arr1)                                      #array([5, 5, 5, 5, 5])
print(arr1*arr2)                                      #array([ 6, 14, 24, 36, 50])
print(arr1/arr2)                                      #array([0.16666667, 0.28571429, 0.375     , 0.44444444, 0.5       ])
print(arr1//arr2)                                     #array([0, 0, 0, 0, 0])
print(arr1%arr2)                                      #array([1, 2, 3, 4, 5])
print(arr1**arr2)                                     #array([      1,     128,    6561,  262144, 9765625])
print(np.power(arr1,arr2))                            #array([      1,     128,    6561,  262144, 9765625])
```

##### RELATIONAL OPERATORS

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([1,2,3,3,5])
print(arr1>arr2)                                  #array([False, False, False,  True, False])
print(arr1>=arr2)                                 #array([ True,  True,  True,  True,  True])
print(arr1<arr2)                                  #array([False, False, False, False, False])
print(arr1<=arr2)                                 #array([ True,  True,  True, False,  True])
print(arr1!=arr2)                                 #array([False, False, False,  True, False])
print(arr1==arr2)                                 #array([ True,  True,  True, False,  True])
```

##### BITWISE OPERATORS

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,9,10])
print("and",arr1&arr2)                    #and [0 2 0 0 0]
print("or",arr1|arr2)                     #or [ 7  7 11 13 15]
print("not",~arr1)                        #not [-2 -3 -4 -5 -6]
print("xor",arr1^arr2)                    #xor [ 7  5 11 13 15]
```

##### LOGICAL OPERATORS

```python
print(np.logical_and(arr1,arr2))           #array([ True,  True,  True,  True,  True])
print(np.logical_or(arr1,arr2))            #array([ True,  True,  True,  True,  True])
print(np.logical_not(arr1))                #array([False, False, False, False, False])
```

```python
# np.sum(op):- It will give sum of all elements in the array
print(np.sum(arr1))                          #np.int64(15) 


# np.mean(op):- It will give the average of all the elements present in a array
print(np.mean(arr1))                         #np.float64(3.0)


# np.max(op):- It will give maximum value among the array
print(np.max(arr1))                          #np.int64(5)


# np.min:- It will give minimum value among the array
print(np.min(arr1))                          #np.int64(1)


# np.count_nonzero(op):- Count of elements
print(np.count_nonzero(arr1))                #5
```
