#### types of array

```python
a=12+4j                  # or we can take 1 or 2 or 3 etc
arr0=np.array(a)
arr0
print(arr0.ndim)                #0-D Array


b=[1,2,3,4]
arr1=np.array(b)
arr1
print(arr1.ndim )               #1-D Array


c=[[1,2,3,4,5,6,7,8]]    # or we can atke   [[1,2,3,4],[5,6,7,8]]
arr2=np.array(c)
arr2
print(arr2.ndim )               #2-D Array


d=[[[1,2,3,4]]]
arr3=np.array(d)
arr3
print(arr3.ndim )               #3-D Array
```

#### to check type of array

```python
a=[1,2,3,4,5]
arr=np.array(a)
print(type(arr))                        # numpy.ndarray



a=[1,2,3,4,5]
arr=np.array(a)
print(type(arr[0]))                     # numpy.int64


a=[1,2,3,4,5.5]
arr=np.array(a)
print(type(arr[0]))                   # numpy.float64


a=[1,2,3,4,5.5,True]
arr=np.array(a)
print(type(arr[0]) )                   # numpy.float64
```
