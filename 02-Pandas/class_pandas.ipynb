```python
import pandas as pd
import numpy as np
```

### 1)Series

```python
# creating a series

ser = pd.Series() 
print("Pandas Series: ", ser,'\n') 

data = np.array(['g', 'e', 'e', 'k', 's']) 
  
ser = pd.Series(data) 
print("Pandas Series:\n", ser)
```

```python
s1=pd.Series([10,20,30,40,'hello',[1,2,4]])
print(s1)
print("***************************************************************************************************")
print(type(s1))
print("***************************************************************************************************")

s1[4]='hi'
s2=pd.Series([1,2,3,4,'world',[9]])
print(s1+s2)
print("***************************************************************************************************")
print(s1)
print("***************************************************************************************************")
print(s2)
print("***************************************************************************************************")
print(s1[4])
```

```python
s3=pd.Series([1,2,3,4,5,6,7,8,9],name="marks")
print(s3)
s4=s3[3:8:2]
print(s4)
```
