```python
d1=pd.DataFrame({'st_name':['pinkiu','harshit','chintu','karan','chintu'],
                  'st_id':[102,103,104,105,104],
                  'st_age':[16,17,18,15,18],
                  'st_ph':[9668193293,9556867980,9732145869,7788993344,9732145869],
                  'st_marks':[99,98,97,96,95]})
d1
```

```python
#change column
# st_ph----> st_phno
# st_mark--->st_score

d1.rename(columns={'st_ph':'st_phno','st_marks':'st_scores'},inplace=True)
d1
```

```python
#By default inplace=false which will not change permanently so we have to assign it True
d1.rename(columns={'st_id':'st_rollno','st_age':'age','st_name':'st'},inplace=True)
d1
```

```python
# it will remove the duplicates from the dataframe which are having same column but temporary
d1.drop_duplicates()
d1
```

```python
# but d1.drop_duplicates() will delete duplicate temporaryly to do it permanant we have to write  inplace=True

d1.drop_duplicates(inplace=True) 
d1 

# or we can write
d1=d1.drop_duplicates()
d1
```

```python
# upd_marks1={98:89,96:69,89:96,78:70}              # by using map if it will not find the key it will show NaN value to overcome it we use replace
# d1['st_scores']=d1['st_scores'].map(upd_marks1)
# d1

"""
output

	st	st_rollno	age	st_phno	st_scores
0	pinkiu	102	16	9668193293	NaN
1	harshit	103	17	9556867980	89.0
2	chintu	104	18	9732145869	NaN
3	karan	105	15	7788993344	69.0
4	chintu	104	18	9732145869	NaN


"""
```

```python
upd_marks2={98:89,96:69,89:96,78:70}                   # it will not show NaN it will replace
d1['st_scores']=d1['st_scores'].replace(upd_marks2)
d1
```

```python
emp=pd.DataFrame({'emp_name':['Kartik', 'Mohan','Sohan', 'Sahil','Karan','Arjun'],
                  'emp_id':[1234,1235,1236,1237,1238,1239],
                  'emp_role':['SE','SE','ASE','JSE','OA','DS'],
                  'emp_years':[2,2,1,1,4,5]})
emp
```

```python
emp.rename(columns={'emp_role':'emp_designation','emp_years':'emp_exp'},inplace=True)    # inplace=True means change permanantly
emp
```

```python
emp['emp_exp']=emp['emp_exp'].replace({1:2})
emp
```

```python
emp['emp_designation']=emp['emp_designation'].map({'SE':'SR.SE','ASE':'SE','JSE':'ASE','OA':'OS','DS':'SRDS'})
emp
```

```python
d3 = pd.read_csv('https://raw.githubusercontent.com/whoknowayush/DataAnalysis-Dataset/main/Datasets-Practice/amazon_fires.csv',encoding='ISO-8859-1')
d3
```

```python
print(d3.info())
d3.rename(columns={'ano':'year','mes':'month','estado':'state','numero':'number_of_fire','encontro':'Date'},inplace=True)
d3
```

```python
# to change or update row value use map function
# inclue all the unique values otherwise it will provide NaN value

month_map = {
    'janeiro': 'January',
    'fevereiro': 'February',
    'março': 'March',
    'abril': 'April',
    'maio': 'May',
    'junho': 'June',
    'julho': 'July',
    'agosto': 'August',
    'setembro': 'September',
    'outubro': 'October',
    'novembro': 'November',
    'dezembro': 'December'
}

d3['month'] = d3['month'].map(month_map)
d3
```
